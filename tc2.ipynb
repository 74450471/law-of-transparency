```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
K = 0.01    # Sensitivity of membrane potential to cognitive activity
B = 0.1     # Baseline rate of change of membrane potential
A = 0.05    # Proportionality constant for current
V = 2.0     # Voltage across the LED
eta = 0.2   # Radiative efficiency of the LED
h = 6.626e-34 # Planck's constant
nu = 5e14   # Frequency of emitted light (example value, in Hz)

# Time array
t = np.linspace(0, 10, 1000)  # 10 seconds, 1000 points

# Simulated cognitive activity in System 1 (example: a sine wave)
A_t = np.sin(2 * np.pi * 0.5 * t)  # 0.5 Hz sine wave

# Function to calculate membrane potential over time
def membrane_potential(A_t, K, B):
    V_m = np.cumsum(K * A_t + B) * (t[1] - t[0])
    return V_m

# Function to calculate current through the LED
def led_current(V_m, A):
    I = A * V_m
    return I

# Function to calculate optical power emitted by the LED
def optical_power(I, V, eta):
    P_electrical = I * V
    P_optical = eta * P_electrical
    return P_optical

# Function to calculate number of photons emitted
def number_of_photons(P_optical, h, nu):
    E_photon = h * nu
    N_photons = P_optical / E_photon
    return N_photons

# Calculate membrane potential
V_m = membrane_potential(A_t, K, B)

# Calculate current through the LED
I = led_current(V_m, A)

# Calculate optical power emitted
P_optical = optical_power(I, V, eta)

# Calculate number of photons emitted
N_photons = number_of_photons(P_optical, h, nu)

# Plot results
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(t, A_t)
plt.title('Cognitive Activity (A_t)')
plt.xlabel('Time (s)')
plt.ylabel('A_t')

plt.subplot(4, 1, 2)
plt.plot(t, V_m)
plt.title('Membrane Potential (V_m)')
plt.xlabel('Time (s)')
plt.ylabel('V_m')

plt.subplot(4, 1, 3)
plt.plot(t, P_optical)
plt.title('Optical Power Emitted (P_optical)')
plt.xlabel('Time (s)')
plt.ylabel('P_optical (W)')

plt.subplot(4, 1, 4)
plt.plot(t, N_photons)
plt.title('Number of Photons Emitted (N_photons)')
plt.xlabel('Time (s)')
plt.ylabel('N_photons')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
K1 = 0.01    # Sensitivity of membrane potential to cognitive activity in System 1
B1 = 0.1     # Baseline rate of change of membrane potential in System 1
A1 = 0.05    # Proportionality constant for current in System 1
V1 = 2.0     # Voltage across the LED in System 1
eta = 0.2    # Radiative efficiency of the LED
h = 6.626e-34 # Planck's constant
nu = 5e14    # Frequency of emitted light (example value, in Hz)

K2_intensity = 0.02  # Sensitivity of System 2 to photon intensity
K2_frequency = 0.01  # Sensitivity of System 2 to photon frequency
B2 = 0.05    # Baseline rate of change of membrane potential in System 2
A2 = 0.1     # Proportionality constant for neural activity response in System 2

# Time array
t = np.linspace(0, 10, 1000)  # 10 seconds, 1000 points

# Simulated cognitive activity in System 1 (example: a sine wave)
A_t1 = np.sin(2 * np.pi * 0.5 * t)  # 0.5 Hz sine wave

# Function to calculate membrane potential over time in System 1
def membrane_potential_system1(A_t, K, B):
    V_m = np.cumsum(K * A_t + B) * (t[1] - t[0])
    return V_m

# Function to calculate current through the LED in System 1
def led_current(V_m, A):
    I = A * V_m
    return I

# Function to calculate optical power emitted by the LED in System 1
def optical_power(I, V, eta):
    P_electrical = I * V
    P_optical = eta * P_electrical
    return P_optical

# Function to calculate number of photons emitted by the LED in System 1
def number_of_photons(P_optical, h, nu):
    E_photon = h * nu
    N_photons = P_optical / E_photon
    return N_photons

# Function to simulate the effect of extra dimensions on photons
def extra_dimensions_effect(N_photons, nu, t):
    # Simulate the effect of extra dimensions as a time-dependent perturbation
    perturbation_intensity = 1 + 0.1 * np.sin(2 * np.pi * 0.1 * t)  # 0.1 Hz modulation
    perturbation_frequency = nu * (1 + 0.05 * np.cos(2 * np.pi * 0.05 * t))  # 0.05 Hz modulation
    
    N_photons_mod = N_photons * perturbation_intensity
    nu_mod = perturbation_frequency
    
    return N_photons_mod, nu_mod

# Function to calculate membrane potential over time in System 2
def membrane_potential_system2(N_photons, nu, K_intensity, K_frequency, B):
    V_m = np.cumsum(K_intensity * N_photons + K_frequency * nu + B) * (t[1] - t[0])
    return V_m

# Function to calculate neural activity in System 2 based on membrane potential
def neural_activity_system2(V_m, A):
    neural_activity = A * V_m
    return neural_activity

# Calculate membrane potential in System 1
V_m1 = membrane_potential_system1(A_t1, K1, B1)

# Calculate current through the LED in System 1
I1 = led_current(V_m1, A1)

# Calculate optical power emitted by the LED in System 1
P_optical1 = optical_power(I1, V1, eta)

# Calculate number of photons emitted by the LED in System 1
N_photons1 = number_of_photons(P_optical1, h, nu)

# Apply extra dimensions effect
N_photons_mod, nu_mod = extra_dimensions_effect(N_photons1, nu, t)

# Calculate membrane potential in System 2 based on modified photons and frequency
V_m2 = membrane_potential_system2(N_photons_mod, nu_mod, K2_intensity, K2_frequency, B2)

# Calculate neural activity in System 2
neural_activity2 = neural_activity_system2(V_m2, A2)

# Plot results
plt.figure(figsize=(12, 12))

plt.subplot(6, 1, 1)
plt.plot(t, A_t1)
plt.title('Cognitive Activity in System 1 (A_t1)')
plt.xlabel('Time (s)')
plt.ylabel('A_t1')

plt.subplot(6, 1, 2)
plt.plot(t, V_m1)
plt.title('Membrane Potential in System 1 (V_m1)')
plt.xlabel('Time (s)')
plt.ylabel('V_m1')

plt.subplot(6, 1, 3)
plt.plot(t, P_optical1)
plt.title('Optical Power Emitted by System 1 (P_optical1)')
plt.xlabel('Time (s)')
plt.ylabel('P_optical1 (W)')

plt.subplot(6, 1, 4)
plt.plot(t, N_photons1)
plt.title('Number of Photons Emitted by System 1 (N_photons1)')
plt.xlabel('Time (s)')
plt.ylabel('N_photons1')

plt.subplot(6, 1, 5)
plt.plot(t, N_photons_mod)
plt.title('Modified Number of Photons Due to Extra Dimensions (N_photons_mod)')
plt.xlabel('Time (s)')
plt.ylabel('N_photons_mod')

plt.subplot(6, 1, 6)
plt.plot(t, neural_activity2)
plt.title('Neural Activity in System 2')
plt.xlabel('Time (s)')
plt.ylabel('Neural Activity')

plt.tight_layout()
plt.show()
```
